```python
#### converion of categorical variables in to dummy variables
## remove car columns
df.drop(['Car_Name'],inplace=True,axis=1)
df.head()
```

```python
final_dataset = pd.get_dummies(df,drop_first=True)
```

```python
sns.heatmap(final_dataset.corr(),annot=True,cmap='YlGnBu')
fig=plt.gcf()
fig.set_size_inches(15,10)
```

```python
### Create Y Variable
Y= final_dataset.Selling_Price
```

```python
final_dataset.drop(['Selling_Price'],inplace=True,axis=1)
```

```python
X=final_dataset
```

```python
### Converting all numerical column in to same scale using min_max scaler
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max = MinMaxScaler()
```

```python
min_max.fit(X)
```

```python
X=pd.DataFrame(min_max.transform(X), columns=X.columns)
```

```python
X.columns
```

```python
### finding important Feature
from sklearn.ensemble import ExtraTreesRegressor
```

```python
model=ExtraTreesRegressor()
model.fit(X,Y)
```

```python
top_features= pd.Series(model.feature_importances_, index=X.columns)
```

```python
top_features.plot(kind='bar')
```

```python
#### Building model 

## Creation of test and train split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
### Hyper parameter tuning
#number of trees in random forest
n_estimators=[int(x) for x in np.linspace(start=100,stop=1200,num=12)]

#number of features to be considered at every split
max_features=['auto','sqrt']

#Max level in tree
max_depth = [int(x) for x in np.linspace(start=5,stop=30,num=6)]

#minimum number of samples required to split the node
min_samples_split=[2,5,10,15,100]

#Minimum number of samples required at each leaf node
min_samples_leaf=[1,2,5,10]
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
## Create RandomGrid
random_grid ={'n_estimators':n_estimators,
              'max_features':max_features,
              'max_depth':max_depth,
              'min_samples_split':min_samples_split,
              'min_samples_leaf':min_samples_leaf
              }
              
print(random_grid)    
            
```

```python
from sklearn.ensemble import RandomForestRegressor 
```

```python
rf=RandomForestRegressor()
```

```python
rf_cv =RandomizedSearchCV(estimator=rf,param_distributions=random_grid,scoring='neg_mean_squared_error',n_iter=10,cv=5,verbose=2,random_state=42)
```

```python
rf_cv.fit(X_train,Y_train)
```

```python
rf_cv.best_params_
```

```python
predictions=rf_cv.predict(X_test)
```

```python
predictions.shape
```

```python
Y_test.shape
```

```python
sns.distplot(Y_test-predictions)
fig= plt.gcf()
fig.set_size_inches(8,8)
```

```python
sns.jointplot(x=predictions,y=Y_test,kind='reg')
```

```python
MSE = np.mean((Y_test-predictions)**2)
print("Mean_Squared_Error",MSE)
```

```python
RMSE = np.sqrt(np.mean((Y_test-predictions)**2))
print("Root_Mean_Squared_Error",RMSE)
```

```python
MAPE= np.mean(np.absolute((Y_test-predictions)/Y_test)*100)
print(MAPE)
```

```python
MAE= np.mean(np.absolute(Y_test-predictions))
print(MAE)
```
